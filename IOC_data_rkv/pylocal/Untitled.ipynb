```python
import fes
import os
import datetime
import numpy as np
```

```python
os.environ['FES_DATA'] = '/home/angel/tools/fes-2.9.1-Source/data/fes2014/ocean_tide'
configuration_file = "/home/angel/tools/fes-2.9.1-Source/data/fes2014/ocean_tide.ini"
```

```python
short_tide = fes.Handler("ocean", "io", configuration_file)
radial_tide = fes.Handler("radial", "io","/home/angel/tools/fes-2.9.1-Source/data/fes2014/load_tide.ini")
```

```python
radial_tide
```

```python
lon = -21.93333333
lat = 64.15
```

```python
date = datetime.datetime(2019, 8, 16)
step = datetime.timedelta(seconds=1800)
```

```python
print("%-27s %9s %9s %9s %7s %7s %8s %8s" % (
        "Date",
        "Latitude",
        "Longitude",
        "Short_tide",
        "LP_tide",
        "Pure_tide",
        "Geo_tide",
        "Rad_tide"))
```

```python
while date < datetime.datetime(2019, 8, 20):
    tide, lp = short_tide.scalar(lat, lon, date)
    if tide is not None:
        load, load_lp = radial_tide.scalar(lat, lon, date)
        print("%-27s %9.3f %9.3f %9.3f %8.3f %9.3f %9.3f %7.3f" % (
            date,
            lat,
            lon,
            tide,
            lp,
            tide + lp, 
            tide + lp + load,
            load))
    date += step
```
